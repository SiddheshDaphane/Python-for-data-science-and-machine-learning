### Basic Matplotlib Commands

```python
plt.plot(x,y,'r') # 'r' is the color red
plt.xlabel('X Axis Title here')
plt.ylabel('Y Axis Title here')
plt.title('Staring Title Here')
plt.show()
```

##### subploting
plt.subplot(nrows,ncols,plot_number)

plt.subplot(1,2,1)
meaning grid of 1 row and 2 columns with last 1 as 1st graph

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(y,x,'g')
plt.subplot(2,2,3)
plt.plot(x,y,'b*--')
plt.subplot(2,2,4)
plt.plot(y,x,'r--')
```

```python
plt.plot(x, np.sin(x), label='sin(x)')
plt.legend()
plt.show()
```
