### Basic Matplotlib Commands

```python
plt.plot(x,y,'r') # 'r' is the color red
plt.xlabel('X Axis Title here')
plt.ylabel('Y Axis Title here')
plt.title('Staring Title Here')
plt.show()
```

##### Multiple Lines

```python
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x,np.cos(x), label='cos(x)')
plt.legend()
plt.show()
```

##### Colors, Linestyles, and Markers

```python
plt.plot(x, np.sin(x), color='red',linestyle='--', marker='o', label='sin(x)')
plt.plot(x, np.cos(x),color='blue', linestyle=':', marker='s', label='cos(x)')
plt.legend()
```

##### Adjusting Axes
Setting specific limits for the x and y axes

```python
plt.plot(x, np.sin(x))
plt.xlim(0,5)
plt.ylim(-1.5,1.5)
```

## Matplotlib Object Oriented Method

```python
# Create Figure (enpty canvas)
fig = plt.figure() 

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # left, bottom, width, height (range 0 to 1)
axes.plot(x,y,'b')
axes.set_xlabel('Set x label')
axes.set_ylabel('Set y label')
axes.set_title('Set Title')
plt.show()
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # inset axes

# Larger Figure axes 1
axes1.plot(x,y,'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure axes 2
axes2.plot(y,x,'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
```

### subplots()

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
for i in axes:
    i.plot(x,y,'b')
    i.set_xlabel('X')
    i.set_ylabel('Y')
    i.set_title('title')

fig
```

**A common issue with matplotlib is overlapping subplots or figures. We can use fig.tight_layout() or plt.tight_layout() method, which automatically adjusts the positions of the axes on the fighure canvas**

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for i in axes:
    i.plot(x,y,'g')
    i.set_xlabel('x')
    i.set_ylabel('y')
    i.set_title('title')

fig
plt.tight_layout()
```

### Figure size, aspect ratio and DPI
**figsize** is a tuple of the width and height of the figure in inches

**dpi** is the dots-per-inch (pixel per inch)

```python
fig = plt.figure(figsize=(8,4), dpi=100)
fig
```

```python
fig, axes = plt.subplots(figsize=(12,3))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

#### Legends

### Setting colors, linewidth, linetypes

```python
# MATLAB style line color and style
fig, ax = plt.subplots()

ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--')
```

#### Colors with the color = parameter

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# possible linestype options -, , -., :, steps
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
